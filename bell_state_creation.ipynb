```python
import qiskit
print("Qiskit version:", qiskit.__version__)
```

```python
from qiskit.visualization import plot_histogram, plot_state_city, plot_bloch_multivector
import matplotlib.pyplot as plt
from qiskit import QuantumCircuit,transpile
from qiskit_aer import AerSimulator
```

## Creation of Bell State $\ket{\Phi^+} = \frac{\ket{00} + \ket{11}}{\sqrt2}$

```python
# Create 2 new qubits and 2 new classical bits
qc1 = QuantumCircuit(2, 2)
# apply H gate to first qubit
qc1.h(0)
# apply CNOT to result of 1st qubit (control) and set result equal to 2nd (target)
qc1.cx(0,1)
# Measure and collapse wavefunction, because system is entangled, measurement of 1st qubit affects 2nd qubit, put results in classical bits
qc1.measure(range(2), range(2))

simulator = AerSimulator()
compiled_circuit = transpile(qc1, simulator)
result = simulator.run(compiled_circuit, shots=1000).result()
counts = result.get_counts()
plot_histogram(counts)

```

```python
qc1.draw("mpl")
```

## Creation of Bell State $\ket{\Phi^-} = \frac{\ket{00} - \ket{11}}{\sqrt2}$

```python
# Create 2 new qubits and 2 new classical bits
qc2 = QuantumCircuit(2, 2)
# apply H gate to first qubit
qc2.h(0)
# apply CNOT to result of 1st qubit (control) and set result equal to 2nd (target)
qc2.cx(0,1)
# flip the sign of 2nd qubit
qc2.z(1)

# Measure and collapse wavefunction, because system is entangled, measurement of 1st qubit affects 2nd qubit, put results in classical bits
# Note: measurement gets rid of the phase, so all measurements are positive
qc2.measure(range(2), range(2))

simulator = AerSimulator()
compiled_circuit = transpile(qc2, simulator)
result = simulator.run(compiled_circuit, shots=1000).result()

counts = result.get_counts()
plot_histogram(counts)
```

```python
qc2.draw("mpl")
```
